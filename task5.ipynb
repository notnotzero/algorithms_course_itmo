```python
g = ig.Graph.Erdos_Renyi(n=100,m=200)
```

```python
ig.plot(g,
        vertex_size=10,
        vertex_label=list(range(200)),
        )
```

```python
print("adjacency matrix \n", g.get_adjacency()[1])
print("adjacency list \n", g.get_adjlist()[1])
```

```python
visited = [False for i in range(101)]
components = []
graph = g.get_adjlist()
def dfs(node, component):
    visited[node] = True
    component.append(node)

    for neighbour in graph[node]:
        if visited[neighbour] == False:
            dfs(neighbour, component)


for i in range(100):
    if visited[i] == False:
        component = []
        dfs(i, component)
        components.append(component)
```
