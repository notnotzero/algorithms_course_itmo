```python
import numpy as np
import matplotlib.pyplot as plt
import scipy

eps = 0.001
```

```python
def f_cube(x):
    return x ** 3

def f_abs(x):
    return abs(x - 0.2)

def f_sin(x):
    return x * np.sin(1 / x)
```

```python
def brute(f, a, b):
    calls = 0
    iterations = 0
    
    n = (b - a) / eps
    interval  = np.linspace(a, b, int(n) + 1)
    min_value = f(interval[0])
    min_arg = interval[0]
    
    calls += 1 
    iterations += 1 # code above is a zero iteration
    
    for x in interval[1:]:
        iterations += 1

        f_value = f(x)
        calls += 1
        
        if f_value < min_value:
            min_value = f_value
            min_arg = x
            
    return min_arg, min_value, calls, iterations
```

```python
min_arg, min_value, calls, iterations = brute(f_cube, 0, 1)
print({'iterations': iterations, 'call functions': calls})

fig, ax = plt.subplots(1, 1, figsize=(15, 10))

x = np.linspace(0, 1, 1000)
y = [f_cube(item) for item in x]


ax.plot(x, 
        y,
        color='blue', 
        label=fr'$x^*={min_arg}, (x^*)^3={min_value}$')

ax.scatter(min_arg, min_value, color='red', s=100)
ax.set_title('Cube brute-force')

ax.set_xlabel('x', fontsize=15)
ax.set_ylabel('$x^3$', fontsize=15)
ax.legend(fontsize=15)
plt.show()

fig.savefig(f'./images/cube_brut_force.png')
```

```python
min_arg, min_value, calls, iterations = brute(f_abs, 0, 1)
print({'iterations': iterations, 'call functions': calls})

fig, ax = plt.subplots(1, 1, figsize=(15, 10))

x = np.linspace(0, 1, 1000)
y = [f_abs(item) for item in x]


ax.plot(x, 
        y,
        color='blue', 
        label=fr'$x^*={min_arg}, f(x^*)={min_value}$')

ax.scatter(min_arg, min_value, color='red', s=100)
ax.set_title('Abs brute-force')

ax.set_xlabel('x', fontsize=15)
ax.set_ylabel(r'$\vert x - 0.2 \vert $', fontsize=15)
ax.legend(fontsize=15)
plt.show()

```
