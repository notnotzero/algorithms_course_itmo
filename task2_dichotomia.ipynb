```python
import numpy as np
import matplotlib.pyplot as plt
import scipy

eps = 0.001
```

```python
def f_cube(x):
    return x ** 3

def f_abs(x):
    return abs(x - 0.2)

def f_sin(x):
    return x * np.sin(1 / x)
```

```python
def dichotomia(f, a, b):
    d = eps / 2
    calls = 0
    iterations = 0
    
    while abs(a - b) > eps:
        x_min = (a + b - d) / 2
        x_max = (a + b + d) / 2

        if f(x_min) <= f(x_max):
            b = x_max
        else:
            a = x_min
        
        iterations += 1
        calls += 2 # calls for left and right x
    
    min_arg = (a + b) / 2
    return min_arg, f(min_arg), calls, iterations
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
x = np.linspace(0, 1, 1000)
y = [f_cube(item) for item in x]
min_arg, min_value, calls, iterations = dichotomia(f_cube, 0, 1)
ax.plot(x, 
        y,
        color='blue', 
        label=fr'$x^*={min_arg}, f(x^*)={min_value}$')
ax.scatter(min_arg, min_value, color='red', s=100)
ax.set_title('Cube dichotomia')
ax.set_xlabel('x', fontsize=15)
ax.set_ylabel('f(x)', fontsize=15)
ax.legend(fontsize=15)
plt.show()
fig.savefig(f'./images/cube_dichotomia.png')

print({'iterations': iterations, 'call functions': calls})
```
