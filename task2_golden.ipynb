```python
import numpy as np
import matplotlib.pyplot as plt
import scipy

eps = 0.001
```

```python
def f_cube(x):
    return x ** 3

def f_abs(x):
    return abs(x - 0.2)

def f_sin(x):
    return x * np.sin(1 / x)
```

```python
def golden_section(f, a, b):
    calls = 0
    iterations = 0
    phi = (np.sqrt(5) + 1) / 2
    
    left = a + (b - a) / phi
    right = b - (b - a) / phi
    value_right = f(right)
    value_left = f(left)
    
    calls += 2
    iterations += 1
    while (b - a) > eps:
        if value_left > value_right:
            b = left
            left = right
            right = b - (b - a) / phi
            value_left = value_right
            value_right = f(right)
        else:
            a = right
            right = left
            left = a + (b - a) /phi
            value_right = value_left
            value_left = f(left)
        
        iterations += 1
        calls+= 1
    
    min_arg = (a + b) / 2
    return min_arg, f(min_arg), calls, iterations
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
x = np.linspace(0, 1, 1000)
y = [f_cube(item) for item in x]
min_arg, min_value, calls, iterations = golden_section(f_cube, 0, 1)
ax.plot(x, 
        y,
        color='blue', 
        label=fr'$x^*={min_arg}, f(x^*)={min_value}$')
ax.scatter(min_arg, min_value, color='red', s=100)
ax.set_title('Cube golden-section')
ax.set_xlabel('x', fontsize=15)
ax.set_ylabel('f(x)', fontsize=15)
ax.legend(fontsize=15)
plt.show()
fig.savefig(f'./images/cube_golden_section.png')

print({'iterations': iterations, 'call functions': calls})
```
